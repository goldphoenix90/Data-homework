```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_data_complete.head()
```

```python
df_by_school = school_data_complete.set_index("school_name")
df_by_school.head()
```

```python
grouped_school_data = df_by_school.groupby("school_name")
grouped_school_data.count().head()
```

```python
score_averages = grouped_school_data["reading_score", "math_score"].mean()
score_averages
```

```python
school_size = grouped_school_data["size"].count()
```

```python
pass_math = df_by_school.loc[df_by_school["math_score"] >=70]
pass_math_total = pass_math.groupby("school_name").count()
percent_math = (pass_math_total["School ID"]/school_size)*100
percent_math
```

```python
pass_read = df_by_school.loc[df_by_school["reading_score"] >=70]
pass_read_total = pass_read.groupby("school_name").count()
percent_read = (pass_read_total["School ID"]/school_size)*100
percent_read
```

```python
overall_passing = (percent_math+percent_read)/2
overall_passing
```

```python
school_budget = grouped_school_data["budget"].mean()
school_budget
```

```python
school_summary = pd.DataFrame(grouped_school_data.count())
school_summary["Average Math Score"] = score_averages["math_score"]
school_summary["Average Reading Score"] = score_averages["reading_score"]
school_summary["budget"] = school_budget
school_summary["Budget Per Student"] = school_summary["budget"]/school_summary["size"]
school_summary["% Passing Math"] = percent_math
school_summary["% Passing Reading"] = percent_read
school_summary["% Overall Passing Rate"] = overall_passing
# school_summary["type"] = school_data["type"]
school_summary_cleaner = school_summary.drop(["type","reading_score", "math_score", "gender", "student_name", "School ID", "Student ID", "grade", ], axis=1)
school_summary_cleaner
```

```python
#Merging with School data to obtain types
school_summary_merged = pd.merge(school_summary_cleaner, school_data, on="school_name")
school_merge_clean = school_summary_merged.drop(["School ID", "size_y", "budget_y"], axis=1)
school_summary_ordered = school_merge_clean[[
    "school_name", "type", "size_x", "budget_x", "Budget Per Student", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"
]]
school_summary_final = school_summary_ordered.rename(columns={
    "school_name": "",
    "type": "School Type",
    "size_x": "Total Students",
    "budget_x": "Total School Budget",
    
})
#Formatting
school_summary_final["Total School Budget"] = school_summary_final["Total School Budget"].map("${:,.2f}".format)
school_summary_final["Budget Per Student"] = school_summary_final["Budget Per Student"].map("${:,.2f}".format)
school_summary_final
```

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

```

```python
city_data.head()
```

```python
ride_data.head()
```

```python
combined_data = pd.merge(ride_data, city_data, on="city")
combined_data.head()
```

```python
average_fare = combined_data.groupby(["city"]).mean()["fare"]
average_fare.head()
```

```python
total_rides = combined_data.groupby(["city"]).count()["ride_id"]
total_rides.head()
```

```python
grouped_data = combined_data.groupby(["type", "city"])
grouped_data.head(10)
```

```python
rides_total = grouped_data.count()["ride_id"]
rides_total.head()
```

```python
mean_fare = grouped_data.mean()["fare"]
mean_fare.head()
```

```python
type_ride_total = combined_data.groupby(["type"]).count()["ride_id"]
percent_rides = (type_ride_total/type_ride_total.sum())*100
percent_rides
```

```python
type_fare_total = combined_data.groupby(["type"]).sum()["fare"]
percent_fares = (type_fare_total/type_fare_total.sum())*100
percent_fares
```

```python
type_driver_total = combined_data.groupby(["type"]).sum()["driver_count"]
percent_drivers = (type_driver_total/type_driver_total.sum())*100
percent_drivers
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_x = rides_total["Urban"]
urban_y = mean_fare.loc["Urban"]

suburban_x = rides_total["Suburban"]
suburban_y = mean_fare.loc["Suburban"]

rural_x = rides_total["Rural"]
rural_y = mean_fare.loc["Rural"]


#Scatter for eachcity type
urban_scatter = plt.scatter(urban_x, urban_y, marker="o", facecolors="coral", edgecolors="black",
            s=urban_x*10, alpha=0.75, label="Urban")
suburban_scatter = plt.scatter(suburban_x, suburban_y, marker="o", facecolors="skyblue", edgecolors="black",
            s=suburban_x*10, alpha=0.75, label="Suburban")
rural_scatter = plt.scatter(rural_x, rural_y, marker="o", facecolors="gold", edgecolors="black",
            s=rural_x*5, alpha=0.75, label = "Rural")
#Formatting scatter plot
plt.legend(loc="best")
plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
# Save Figure
plt.savefig("../Images/PyberScatter2018.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_percent = percent_fares
city_types = ["Rural", "Suburban", "Urban"]
type_colors = ["Gold", "SkyBlue", "Coral"]
explode = (0,0,0.2)

# Build Pie Chart
plt.pie(type_percent, explode=explode, labels=city_types, colors=type_colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/PyberPieFares2018.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
type_percent = percent_rides
city_types = ["Rural", "Suburban", "Urban"]
type_colors = ["Gold", "SkyBlue", "Coral"]
explode = (0,0,0.2)

# Build Pie Chart
plt.pie(type_percent, explode=explode, labels=city_types, colors=type_colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Images/PyberPieRides2018.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
type_percent = percent_drivers
city_types = ["Rural", "Suburban", "Urban"]
type_colors = ["Gold", "SkyBlue", "Coral"]
explode = (0,0,0.2)

# Build Pie Chart
plt.pie(type_percent, explode=explode, labels=city_types, colors=type_colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/PyberPieDrivers2018.png")
plt.show()
```
